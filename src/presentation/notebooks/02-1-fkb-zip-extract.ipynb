```python
from io import BytesIO

import requests
import geopandas as gpd

from src.infra.infrastructure.services.zip_service import ZipService
```

```python
zip_service = ZipService()
```

```python
paths = [
    "https://huggingface.co/datasets/kartai/DX_datasett/resolve/main/Geodata/Bergen.zip",
    "https://huggingface.co/datasets/kartai/DX_datasett/resolve/main/Geodata/Kristiansand.zip",
    # "https://huggingface.co/datasets/kartai/DX_datasett/resolve/main/Geodata/Mo_i_Rana.zip",
    "https://huggingface.co/datasets/kartai/DX_datasett/resolve/main/Geodata/Sandvika.zip",
    # "https://huggingface.co/datasets/kartai/DX_datasett/resolve/main/Geodata/Tromsdalen.zip",
    "https://huggingface.co/datasets/kartai/DX_datasett/resolve/main/Geodata/Verdal.zip"
]
token = "token"
headers = {"Authorization": f"Bearer {token}"}
layer_bytes: list[BytesIO] = []

for path in paths:
    response = requests.get(path, headers=headers, stream=True)
    response.raise_for_status()

    zip_bytes = BytesIO(response.content)
    layers = ["Bygning", "AnnenBygning", "Takkant", "Bygningsdelelinje", "FiktivBygningsavgrensning"]
    bytes = zip_service.unzip_flat_geobuf(zip_bytes, *layers)
    layer_bytes.extend(bytes)
```

```python
layer_bytes
```

```python
gdfs: list[gpd.GeoDataFrame] = []
for layer in layer_bytes:
    gdf = gpd.read_file(layer, engine="pyogrio")
    gdfs.append(gdf)

merged = gpd.pd.concat(gdfs, ignore_index=True)
wgs_84 = merged.to_crs(4326)
wgs_84.to_file("data.json")
```
